### 1) What is the relationship between sensation seeking and movie experience? 🤔

* Perform a correlation analysis between these averages.
* Create a scatter plot to visualize the relationship.

### 3) Are movies that are more popular rated higher than movies that are less popular 🔝?

**📌 Greater Objective:** Determine if popularity causes higher ratings, vice versa.

**📝 What I will do:**
* Calculate the popularity of each movie (e.g., by the number of ratings).
* Calculate the average rating for each movie.
* Analyze the correlation between popularity and average rating.

```python
# Calculate movie popularity and average rating
movie_data = df.iloc[:, :400]
movie_popularity = movie_data.notna().sum()
movie_avg_rating = movie_data.mean()
```

```python
# Create a dataframe with popularity and average rating
movie_analysis = pd.DataFrame({
    'Popularity': movie_popularity,
    'Average Rating': movie_avg_rating
})
```

```python
# Calculate correlation
correlation, p_value = stats.pearsonr(movie_analysis['Popularity'], movie_analysis['Average Rating'])
```

```python
# Define exponential function to fit to represent the trend
def exp_func(x, a, b):
    return a * np.exp(b * x)

# Fit the exponential function to the data
params, params_covariance = curve_fit(exp_func, movie_analysis['Popularity'], movie_analysis['Average Rating'], p0=[2, 0.01])

# Create scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Popularity', y='Average Rating', data=movie_analysis, color='dodgerblue')
plt.title('Movie Popularity vs Average Rating')
plt.xlabel('Number of Ratings (Popularity)')
plt.ylabel('Average Rating')

# Plot the fitted exponential trend line
x_values = np.linspace(0, movie_analysis['Popularity'].max(), 400)
plt.plot(x_values, exp_func(x_values, *params), color='red', label='Fitted exponential line')

plt.legend()
plt.show()
```

```python
print(f"Correlation: {correlation:.5f}, p-value: {p_value:.5e}")
```

**Analyzing the Graph 📊:**
* The scatter plot shows a clear positive trend, which is emphasized by the fitted exponential line, indicating that as the number of ratings (popularity) increases, so does the average rating of a movie.

* The data points are more concentrated at the lower end of the popularity scale (fewer ratings), with ratings generally increasing but showing a wider variation as popularity increases.

**Analyzing the Stats 💡:**
* A Correlation Coefficient of 0.699 is relatively high, indicating a substantial positive correlation between the number of ratings and the average rating. This suggests that movies with more ratings tend to have higher average ratings.

* A P-value of 5.769e-60 is extremely low, which indicates that the observed correlation is highly statistically significant and there is basically 0 chance that this correlation occurred by chance.

**My Thoughts 💭:**
* The trend suggests that movies which attract more viewers are generally better received, possibly because they are of higher quality, more engaging, or have better marketing.
* There could be a bias where popular movies receive higher ratings simply because they are more likely to be rated by fans or because viewers may have higher expectations and a readiness to rate them favorably.
* The dataset might be subject to selection bias. For example, if only blockbusters tend to get more ratings, they might inherently have qualities that lead to higher ratings.
* Movies with very few ratings might suffer from extreme ratings which don't accurately reflect general viewer sentiment as compared to popular movies where the average rating is more likely to converge towards a truer mean.

**Conclusion 🙌:**
* The analysis suggests a strong positive relationship between movie popularity and average ratings, affirming that more popular movies are, on average, rated higher than less popular ones. 
* Further analysis would is needed to understand the causative factors behind this trend.

### 4) Is enjoyment of 'Shrek (2001)' gendered, i.e. do male and female viewers rate it differently?

**📌 Greater Objective:** To start testing if gender is a defining predictor trait.

**📝 What I will do:**
* Extract Shrek ratings and gender data.
* Removes non-binary responses for simplicity.
* Performs an independent t-test to compare ratings between genders.
* Creates a box plot to visualize the distribution of ratings by gender.

```python
# Get the Shrek Ratings
shrek_column = df.columns.get_loc("Shrek (2001)")
shrek_ratings = df.iloc[:, shrek_column]
gender = df.iloc[:, 474]
```

```python
# Create a DataFrame with Shrek Ratings+Gender Data
shrek_data = pd.DataFrame({'Rating': shrek_ratings, 'Gender': gender})

# Check for and drop any NaN values in 'Rating' or 'Gender'
shrek_data = shrek_data.dropna(subset=['Rating', 'Gender'])

# Remove any non-binary gender responses for this analysis
shrek_data = shrek_data[(shrek_data['Gender'] == 1) | (shrek_data['Gender'] == 2)]
```

```python
# Perform t-test
female_ratings = shrek_data[shrek_data['Gender'] == 1]['Rating']
male_ratings = shrek_data[shrek_data['Gender'] == 2]['Rating']
t_stat, p_value = stats.ttest_ind(female_ratings, male_ratings, nan_policy='omit')  # Handling NaN values just in case
```

```python
# Visualize the data
plt.figure(figsize=(10, 6))

ax = sns.boxplot(x='Gender', 
                 y='Rating', 
                 data=shrek_data,
                 palette=["#F06292", "#64B5F6"])

ax.set_xticklabels(['Female', 'Male'])
ax.set_ylabel('Rating', fontsize=16)
ax.set_xlabel('Gender', fontsize=16) 
plt.title("Shrek (2001) Ratings by Gender", fontsize=18)

plt.show()
```

```python
print(f"T-statistic: {t_stat:.5f}, p-value: {p_value:.5f}")
```

**Analysis 💡:**
* The positive t-stat indicates that men might like Shrek more than women; however the p-value is not below a a significant level of even .05, so there is no statistically significant difference of the enjoyment of Shrek based on gender 🤝.

### 7) There are ratings on movies from several franchises 🦖 (E.G., Star Wars, Batman, Toy Story) in this dataset 🦇. How many of these are of inconsistent quality, as experienced by viewers? 😕


**📌 Greater Objective:** Identify inconsistencies in viewer ratings across selected movie franchises.


**📝 What I will do:**
* Find the exact name of each franchises' movies.
* Find the standard deviation of ratings within each franchise
* Performs an independent t-test to compare the average franchise consistency to overall movies.
* Creates a bar plot to visualize the distribution of standard deviations by franchise.
* Create subplots that show the varied consistency of movie franchises.

```python
# Algorithm I used to find the exact string names for each franchise
search_string = 'Star Wars'
matching_columns = [col for col in df.columns if search_string in col]
matching_columns
```

```python
# List of the franchises
franchises = {
    'Star Wars': ['Star Wars: Episode IV - A New Hope (1977)',
                  'Star Wars: Episode V - The Empire Strikes Back (1980)',
                  'Star Wars: Episode VI - The Return of the Jedi (1983)',
                  'Star Wars: Episode 1 - The Phantom Menace (1999)',
                  'Star Wars: Episode II - Attack of the Clones (2002)',
                  'Star Wars: Episode VII - The Force Awakens (2015)'],
    
    'Harry Potter': ["Harry Potter and the Sorcerer's Stone (2001)",
                     'Harry Potter and the Chamber of Secrets (2002)',
                     'Harry Potter and the Goblet of Fire (2005)',
                     'Harry Potter and the Deathly Hallows: Part 2 (2011)'], 
    
    'The Matrix':['The Matrix (1999)', 
                  'The Matrix Revolutions (2003)',
                  'The Matrix Reloaded (2003)'],
    
    'Indiana Jones':['Indiana Jones and the Raiders of the Lost Ark (1981)',
                     'Indiana Jones and the Temple of Doom (1984)', 
                     'Indiana Jones and the Last Crusade (1989)',
                     'Indiana Jones and the Kingdom of the Crystal Skull (2008)'], 
    
    'Jurassic Park':[ 'Jurassic Park (1993)', 
                     'The Lost World: Jurassic Park (1997)',
                     'Jurassic Park III (2001)'], 
    
    'Pirates of the Caribbean':['Pirates of the Caribbean: The Curse of the Black Pearl (2003)', 
                                "Pirates of the Caribbean: Dead Man's Chest (2006)",
                                "Pirates of the Caribbean: At World's End (2007)"],
    
    'Toy Story':['Toy Story (1995)', 
                 'Toy Story 2 (1999)', 
                 'Toy Story 3 (2010)'], 
    
    'Batman':['Batman (1989)', 
              'Batman & Robin (1997)', 
              'Batman: The Dark Knight (2008)']
}
```

```python
# Analyze the standard deviation of ratings within each franchise
variability = {}
for franchise, movies in franchises.items():
    franchise_data = df[movies]
    variability[franchise] = franchise_data.std(axis=0).mean()
```

```python
# Get the inconsistency of all movies combined
movie_ratings = df.iloc[:, :400]
movie_std_dev = movie_ratings.std()
average_std_dev = movie_std_dev.mean()

# Add to the variability list
variability["All Movies"] = average_std_dev
```

```python
# Vizualization
colors = ['#E57373', '#F06292', '#BA68C8', '#64B5F6', '#4DB6AC', '#FFD54F', '#FF8A65', '#FFAB91', '#A1887F']

plt.bar(variability.keys(), variability.values(), color=colors)
plt.ylabel('Average Standard Deviation of Ratings')
plt.ylim(0, 1.2)
plt.title('Inconsistency in Movie Franchises')
plt.xticks(rotation=45, ha='right')

plt.show()
```

**Analysis of the Graph 📍:**
* Just looking at franchise averages compared to all movies, we can see that viewers tend to have much less varied views when a movie is in a franchise then when not 👍.
* Although this conclusion may be intuitive, movies where the public generally agrees on it being good means it is probably going to get a remake, constituting a franchise 👏.
* There can also be some larger biases at hand 🙅🏻; for example, a franchise is likely much more established than a new IP, so viewers may have a stronger opinion about a franchised film before even going to see it.

```python
# Initialize a dictionary to store results
franchise_std_devs = {}

# Loop through each franchise and calculate standard deviations
for franchise, movies in franchises.items():
    selected_movies = df[movies]
    std_devs = selected_movies.std(axis=0)
    franchise_std_devs[franchise] = std_devs
```

```python
# Plotting the standard deviations to visually compare inconsistency within franchises
fig, ax = plt.subplots(figsize=(10, 6))
for franchise, std_devs in franchise_std_devs.items():
    ax.plot(std_devs, marker='o', label=franchise)

ax.set_ylabel('Standard Deviation of Ratings')
ax.set_title('Inconsistency within Movie Franchises')
ax.legend(title='Franchise', loc='upper left', bbox_to_anchor=(1, 1))
ax.set_xticks([])
plt.tight_layout()

plt.show()
```

**Analysis 💡:** 
* Franchises with relatively flat lines (e.g., Toy Story 🚂) indicate consistency in movie quality and audience reception, which might be indicative of a successful formula that resonates well with its audience across different movies 🎉.
* The franchises with higher peaks and more pronounced fluctuations (e.g., Star Wars 🛸, The Matrix) suggest that some movies were much more controversial or received mixed reviews 💔 compared to others within the same franchise.
* Overall, this graph highlights the variability in movie ratings that exists within franchises, with most franchises having varying levels of fan reception 🎙️.
* The franchises with greater inconsistency might benefit from a deeper dive to understand the causes of divergent view points 🔎; this could be related to changes in direction, tone, or quality across the series.

```python
# Extract movie names from franchises
franchise_movies = [movie for movies in franchises.values() for movie in movies]

# Calculate standard deviations for franchise movies
franchise_stds = df[franchise_movies].std()

# Calculate standard deviations for all movies
all_movie_stds = df.iloc[:, :400].std()  # assuming movie rating columns are the first 400 columns

# Two-sample t-test
t_stat, p_value = ttest_ind(franchise_stds, all_movie_stds, equal_var=False)

print(f'T-statistic: {t_stat:.5f}, P-value: {p_value:.5e}')
```

**Consistency Conclusion 🚀:**
* The negative sign of the T-statistic indicates that the mean standard deviation of the franchise movies is lower than that of all movies. This suggests that movies within these franchises tend to have more consistent ratings compared to the general pool of movies. The magnitude 🌋 of the T-statistic, which is quite large, highlights that this is a notable difference.
* The P-value is extremely small 🤏, far less than the commonly used significance threshold of 0.05. This result implies that the observed difference in standard deviations is highly unlikely to have occurred by chance. This means the franchises I analyzed have significantly more consistent ratings (higher agreement) than movies in general 🤝🏽. 
* There is a higher level of agreement among viewers about the quality of movies within these franchises 💯, or could reflect the effectiveness of these franchises in maintaining a consistent quality.
* For stakeholders in the movie industry 🧑🏾‍💼, these findings underscore the value of franchise films in terms of predictability in audience reception. This could potentially justify investment in franchise films due to their reliable audience appeal. Additionally, these results could guide marketing strategies by promoting materials 📢 that attract audiences seeking a reliable entertainment experience!

### 8) Build a prediction model (supervised learning) to predict movie ratings (for all 400 movies!) based only on personality traits. I will also use cross-validation methods to avoid overfitting and characterize the accuracy of my model.

**📌 Greater Objective:** Find out which personality traits are the most important to predict the enjoyment of a movie.

**📝 What I will do:**
* Extracting and preparing the relevant data.
* Leveraging a Random Forest Regressor to implement 5-fold cross-validation, shuffle the data, and create a fixed random state.
* Fit model using personality traits and average movie.
* Calculate cross-validation scores using negative mean squared error to find the Root Mean Squared Error (RMSE) and accuracy of the model.
* Plot the relative importance of each personality factor in predicting movie ratings.
* Plot predicted movie ratings against actual ratings to visualize model performance.

```python
# Prepare data
X = df.iloc[:, 421:464]  # Personality factors
y = df.iloc[:, :400].mean(axis=1)  # Average movie rating per user

# Handling missing data
X.fillna(X.mean(), inplace=True)
y.fillna(y.mean(), inplace=True)
```

```python
# Set up the Model
model = RandomForestRegressor(n_estimators=100)
kf = KFold(n_splits=5, shuffle=True, random_state=42)
```

```python
# Cross-validation
scores = cross_val_score(model, X, y, cv=kf, scoring='neg_mean_squared_error')
rmse = np.sqrt(-scores.mean())
print("Average RMSE:", np.sqrt(-scores.mean()))
```

```python
# Calculate the accuracy as a percentage
rating_range = 4.0
normalized_rmse = rmse / rating_range # Calculate normalized RMSE
accuracy = (1 - normalized_rmse) * 100 # Calculate accuracy

print("Accuracy:", accuracy, "%")
```

**🎯 RMSE and Accuracy Analysis:**

```python
# Assuming PCA and scaling have already been performed
pca = PCA(n_components=50)
pca.fit(X_all_scaled)  # Fit PCA to the scaled data

# Plotting the variance explained by each PCA component
plt.figure(figsize=(10, 6))
plt.bar(range(1, 51), pca.explained_variance_ratio_, alpha=0.5, align='center', label='individual explained variance')
plt.step(range(1, 51), np.cumsum(pca.explained_variance_ratio_), where='mid', label='cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.title('PCA Variance Explained')
plt.show()
```

**Analysis 💡:**
* This plot shows how much of the total variance is captured by each principal component 🌌. It's useful for understanding the effectiveness of PCA in reducing dimensionality while retaining information.
* For example, the first 3 PCIs, or dimensions, contain 30% of all the info gained from these 77 questions 🌪️; the first 10 contain 50% of all of the info. 
* 3/77 and 10/77 is a significant reduction in the total dimensions. If the most impactful questions don't have too much info loss in the following steps, they can potentially be great predictors of movie ratings 📐.

```python
# Get feature importances
importances = full_model.feature_importances_
indices = np.argsort(importances)[::-1]

# Since we used PCA, we don't have original feature names, but if we want to show PCA components:
components = [f'PC{i+1}' for i in range(X_pca.shape[1])]  # Create labels for PCA components

# Plot
plt.figure(figsize=(12, 8))
plt.title('Feature Importances by PCA Components', fontsize=18)
plt.bar(range(X_pca.shape[1]), importances[indices], color='#E57373', align='center')
plt.xticks(range(X_pca.shape[1]), np.array(components)[indices], rotation=90, fontsize=12)
plt.xlim([-1, X_pca.shape[1]])
plt.xlabel('PCA Component', fontsize=16)
plt.ylabel('Relative Importance', fontsize=16)
plt.tight_layout()
plt.show()
```

**Analysis 💡:**
